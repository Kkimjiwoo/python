```python
# 데이터프레임 결측치 처리 

df['age_na'] = np.nan
df.head()
```

```python
# 결측치를 0으로 채움
df['age_na'].fillna(0)
```

```python
df['age_na'] = df['age_na'].fillna(df['housing_median_age'].mean())
df.head(3)
# 'age_na' 열을 housing_median_age 열의 평균으로 채운다
```
