```python
# 그래프 그려보기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.title("My first Plot")
plt.xlabel('x')
plt.ylabel('y')
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

fig, ax = plt.subplots()
# fig 변수는 생성된 figure를 나타내며, 이 figure는 하나 이상의 subplot을 포함
# ax 변수는 생성된 subplot을 나타냅니다. subplot은 그래프를 그리는 영역이며, 
# 이 경우엔 하나의 subplot이 생성
ax.plot(x,y)
ax.set_title('First Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')

fig.set_dpi(300)    # 해상도를 높여줌
fig.savefig("first_plot.png")   # 그림 저장
# fig : figure 그림
plt.show()
```

```python
# 여러 개 그래프 그리기

x = np.linspace(0, np.pi*4, 100)    # 100등분
fig, axes = plt.subplots(2,1)   # 2행 1열 두개의 그림

axes[0].plot(x,np.sin(x))
axes[1].plot(x,np.cos(x))

plt.show()
```

Matplotlib 주요 그래프

```python
# 선 그래프 (line 그래프)

fig, ax = plt.subplots()

x = np.arange(15)
y = x**2 
ax.plot(
    x,y,
    linestyle = ":",
    marker = "*",
    color = "darkblue"
)

plt.show()
```

```python
# color

x = np.arange(10)

fig, ax = plt.subplots()

ax.plot(x,x,color = 'r') # color = 'r' >> red
ax.plot(x,x+2,color = 'blue')
ax.plot(x,x+4,color = '0.8')    # 0.8은 연한 회색(명도 조절)
# 그레이스케일 (흑백) 0~1 : 검정색, 1 : 하얀색
ax.plot(x,x+6,color = '#524FA1')

plt.show()
```

```python
# Marker 

x = np.arange(10)
fig, ax = plt.subplots()
ax.plot(x, x, marker = '.')
ax.plot(x, x+2, marker = 'o')
ax.plot(x, x+4, marker = 'v')
ax.plot(x, x+6, marker = 's')
ax.plot(x, x+8, marker = '*')

plt.show()
```

```python
# 축 경계 조정하기
x = np.linspace(0,10,1000)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-2,12)      # x축 경계
ax.set_ylim(-1.5,1.5)   # y축 경계
```

```python
# 범례

fig, ax = plt.subplots()
ax.plot(x,x,label='y=x')
ax.plot(x,x**2, label = 'y=x^2')

ax.set_title("legend of graph")
ax.set_xlabel("x")
ax.set_ylabel("y")

ax.legend(loc = 'best')
          #shadow=True,
          #fancybox = True, # 둥글게
          #borderpad=2)
          # 범례의 위치를 자동으로 최적화 설정
          
plt.show()
```

```python
# 산점도 (Scatter)

fig, ax = plt.subplots()

x = np.arange(10)
ax.plot(
    x,x**2,"o",
    markersize = 15,
    markerfacecolor = "white",
    markeredgecolor = "blue"
)

plt.show()
```

```python
# # scatter
fig, ax = plt.subplots()

x = np.arange(10)
ax.scatter(
  x, x**2
)
plt.show()
```

```python
# bubble 형태 scatter

fig, ax = plt.subplots()

x = np.random.randn(50)
y = np.random.randn(50)

colors = np.random.randint(0,100,50)
sizes = 500 * np.pi * np.random.rand(50) **2
# 사이즈의 길이가 50이므로 총 50개의 크기가 존재

ax.scatter(x,y,c = colors, s=sizes, alpha = 0.3 )
# x,y는 축데 대한 데이터 지정
# c는 컬러
# s는 데이터 포인트의 크기를 지정(위에 지정함)
# alpha는 투명도(0~1 사이)

plt.show()
```

```python
# 바 그래프 bar plot

# bar
x = np.arange(10)

fig, ax = plt.subplots(figsize=(12,4))  # (12,4)인치로 나옴
ax.bar(x,x**2)

plt.show()
```

```python
# bar plot 누적 바 그래프

x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)

data = [x,y,z]

fig, ax = plt.subplots()
x_ax = np.arange(3)

for i in x_ax:
    ax.bar(x_ax, data[i],
          bottom=np.sum(data[:i], axis =0))
    
ax.set_xticks(x_ax)
ax.set_xticklabels(["A","B","C"])    
```

```python
# 히스토그램 (HISTOGRAM)
# 연속형 데이터
fig, ax = plt.subplots()
data = np.random.randn(1000)
# 평균이 0이고 표준편차가 1인 정규분포를 따름

ax.hist(data, bins=50)  # 계급구간 bins
plt.show()
```

```python
path = '../data/pokemon.csv'
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df.columns
```

```python
# condition 조건
condition_01 = (df['Type 1'] == 'Fire') | (df['Type 2'] == 'Fire')
df[condition_01]
# '|' 둘 중 하나라도 참인 경우 
```

```python
fire = df[condition_01]
fire
```

```python
condition_02 = (df['Type 1'] == 'Water') | (df['Type 2'] == 'Water')
df[condition_02]
```

```python
water = df[condition_02]
```

```python
fig, ax = plt.subplots()
ax.scatter(fire['Attack'], fire['Defense'],
           color='r', label='Fire', marker='*', s=50)

ax.scatter(water['Attack'], water['Defense'],
           color='blue', label='water', marker='o', s=50)

ax.set_xlabel('Attack')
ax.set_ylabel('Defense')
ax.legend(loc='best')

plt.show()
```
