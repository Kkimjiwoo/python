```python
from collections import Counter

# 관광 유형 리스트들을 하나의 리스트로 합침
all_tourism_types = nature_tourism + historical_tourism + experience_tourism + theme_tourism + cultural_tourism + shopping_and_others

# 원소 등장 횟수를 세기 위해 Counter 객체 생성
tourism_counter = Counter(all_tourism_types)

# 등장 횟수가 2 이상인 원소 출력
overlap_elements = [element for element, count in tourism_counter.items() if count > 1]

if overlap_elements:
    print("겹치는 원소:", overlap_elements)
else:
    print("겹치는 원소가 없습니다.")
```

```python
# 해당 관광 데이터 프레임 만들기

def tourism_category(tourism):  # 이때 자연, 역사, 체험, 테마 중 선택
    if tourism == '자연':
        nature_info_tourism = all_tourism_info[all_tourism_info['Subclass'].isin(nature_tourism)]
        info = nature_info_tourism
    elif tourism == '역사':
        historical_info_tourism = all_tourism_info[all_tourism_info['Subclass'].isin(historical_tourism)]
        info = historical_info_tourism        
    elif tourism == '체험':
        experience_info_tourism = all_tourism_info[all_tourism_info['Subclass'].isin(experience_tourism)]
        info = experience_info_tourism        
    elif tourism == '테마':
        theme_tourism_info = all_tourism_info[all_tourism_info['Subclass'].isin(theme_tourism)]
        info = theme_tourism_info
    elif tourism == '문화':
        cultural_tourism_info = all_tourism_info[all_tourism_info['Subclass'].isin(cultural_tourism)]
        info = cultural_tourism_info
    elif tourism == '쇼핑 및 기타':
        shopping_and_others_info_tourism = all_tourism_info[all_tourism_info['Subclass'].isin(shopping_and_others)]
        info = shopping_and_others_info_tourism       
    return info 
        
        	
```

```python
nature_review_categories = {
    "자연 경관/경관 관련": ["신기한 식물이 많아요", "뷰가 좋아요", "야경이 멋져요", "경관이 독특해요", "자연 경관이 좋아요"],
    "시설/편의성 관련": ["산책로가 잘 되어있어요", "화장실이 깨끗해요", "편의시설이 잘 되어있어요", "주차하기 편해요", "대중교통이 편해요", "시설이 깔끔해요", "관리가 잘 되어있어요", "인테리어가 멋져요"],
    "가격 관련": ["가격이 합리적이에요"],
    "가족/어린이 관련": ["아이와 가기 좋아요", "피크닉하기 좋아요", "반려동물과 가기 좋아요", "놀이기구가 다양해요"],
    "활동/체험 관련": ["볼거리가 많아요", "체험 프로그램이 다양해요"],
    "환경/분위기 관련": ["붐비지 않아요", "조용히 쉬기 좋아요", "유익해요", "친절해요", "트렌디해요", "컨셉이 독특해요", "인테리어가 멋져요"],
    "기타": ["규모가 커요", "방문객이 많아요", "코스가 길어요", "동물 관리가 잘 되어있어요", "대기시간이 짧아요", "먹거리가 풍부해요",'사진이 잘 나와요']
}

# 데이터프레임 생성
df = pd.DataFrame.from_dict(nature_review_categories, orient='index').transpose()

# 결과 출력
df
```

```python
nature = tourism_category('자연')
nature[:2]
```

```python
# Good_Point_categories 컬럼 초기화
Good_Point_categories = []

# 각 Good_Point가 데이터프레임에 있는지 확인하고, 있다면 카테고리를 Good_Point_categories 리스트에 저장
for Good_Point in nature['Good_Point']:
    categories_found = []
    for category in df.columns:
        if Good_Point in df[category].values:
            categories_found.append(category)
    Good_Point_categories.append(', '.join(categories_found))

# 새로운 열로 추가
nature['Good_Point_categories'] = Good_Point_categories
```

```python
sum_category =nature.groupby(['Attraction','Good_Point_categories'])['Selected_People'].sum().reset_index()
sum_category.rename(columns = {	'Selected_People':'sum_category_Selected_People'}, inplace = True)
```

```python
sum_selected_People =nature.groupby(['Attraction'])['Selected_People'].sum().reset_index()
sum_selected_People.rename(columns = {'Selected_People':'sum_selected_People'}, inplace = True)
```

```python
df_nature = pd.merge(nature,sum_category , on=['Attraction','Good_Point_categories'])
df_nature = pd.merge(df_nature,sum_selected_People, on=['Attraction'])
df_nature['one_attraction_selected_ratio'] = df_nature['sum_selected_People']/df_nature['Participants']
```
